This machine learning project concerns data acquired by submaries to identify if there are any mines in the water. There are 2 possible outputs, it's eithr a rock or a mine.

## Importing libraries for the project

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#import dataset and skip header since there are no names assigned to columns in dataset
df = pd.read_csv('Copy of sonar data.csv', header=None)
df
```

```python
#208 rows / 61 columns
df.shape
```

```python
#Statistical data
#Here we can see that there are no missing values in the dataframe
df.describe()
```

```python
#Let's rename the independent variable
df.rename(columns={60: 'Rock/Mine'},inplace=True)
df
```

```python
#We have 111 mines and 97 rocks in the dataset
#The data is quite balanced and it should give good predictions and accuracy for the model
df['Rock/Mine'].value_counts()
```

```python
df.groupby('Rock/Mine').mean()
```

```python
#X reprensents the features or the independent variables
#Y represents the output or the dependent variable
X = df.drop(columns = 'Rock/Mine', axis=1) #dependent variables or features
Y = df['Rock/Mine'] #independent variable, rock or mine, binary
```

```python
print(X)
print(Y)
```

## Train / Test splitting

```python
#Test size here is 10%
#stratify keeps same percentage of classes in each split
#Random state is used if we want to get the same sample if someone wants to do the same test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.1, stratify=Y, random_state = 1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model training

```python
# model choice => Logistic regression which is a classification model
model = LogisticRegression()
```

```python
#Let's train the model in training data
model.fit(X_train, Y_train)
```

```python
#Evaluation of the model and accuracy on training data
X_prediction = model.predict(X_train)
training_accuracy = accuracy_score(X_prediction, Y_train)
```

```python
print("The accuracy of the training data is: ", training_accuracy*100 , "%")
```

```python
#Evaluation of the model and accuracy on testing data
X_test_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("The accuracy of the testing data is: ", testing_accuracy*100 , "%")
```

##  Support Vector Classifier model

```python
#SVC is also a machine learning algorithm used for classification purposes
#It finds the optimal hyperplane the divides the dataset into 2 classes
#For this dataset SVC gave us a better accuracy that LogisticRegression
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, Y_train)
X_svc_predict = model.predict(X_test)
accuracy = accuracy_score(X_svc_predict,Y_test)
print("SVC accuracy = ",accuracy * 100,"%")
```
